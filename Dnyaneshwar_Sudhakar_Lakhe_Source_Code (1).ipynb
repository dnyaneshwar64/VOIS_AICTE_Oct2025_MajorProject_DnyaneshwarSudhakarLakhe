```python
import pandas as pd
import numpy as np
import re

def show_cols(df, title="Columns"):
    """Helper function to display column names clearly."""
    print(f"\n--- {title} ({len(df.columns)}) ---")
    for c in df.columns:
        print(repr(c))
    print('--------------------------\n')

# --- Main Cleaning Process ---
print("Starting Netflix data cleaning pipeline...")

try:
    df = pd.read_csv("Netflix Dataset.csv")
    print(f"Successfully loaded 'Netflix Dataset.csv'. Shape: {df.shape}")
except FileNotFoundError:
    print("ERROR: 'Netflix Dataset.csv' not found. Please place it in the same folder.")
    exit()

# 1) Inspect raw columns (shows whitespace or odd characters)
show_cols(df, "RAW columns (before normalization)")

# 2) Normalize: lowercase, strip, collapse whitespace -> underscore
# This will turn 'Show_Id' into 'show_id' and 'Release_Date' into 'release_date'
new_cols = {}
for c in df.columns:
    clean = re.sub(r'\s+', ' ', str(c)).strip().lower().replace(' ', '_').replace('-', '_')
    new_cols[c] = clean

df.rename(columns=new_cols, inplace=True)
show_cols(df, "Normalized columns (lower/strip/underscores)")

# 3) Map common alternate names to canonical names
# We adapt this for our Netflix data
alias_map = {
    'show_id': 'show_id',
    'id': 'show_id',      # 'id' is a common alias for 'show_id'
    'type': 'genres',     # The 'Type' column actually contains genres
    'category': 'media_type' # 'Category' is 'Movie' or 'TV Show'
}

# Rename present alias keys to canonical names if necessary
rename_to = {}
for col in df.columns:
    for alt, std in alias_map.items():
        if col == alt and col != std:
            rename_to[col] = std

if rename_to:
    df.rename(columns=rename_to, inplace=True)
    print("Applied alias renames:", rename_to)
show_cols(df, "After alias mapping")

# 4) Check for required columns / candidates
# We change this list to what *we* need for our analysis
required = ['show_id', 'media_type', 'title', 'country', 'release_date', 'rating', 'duration', 'genres']
for r in required:
    if r not in df.columns:
        candidates = [c for c in df.columns if r in c or r in c.replace('_','')]
        print(f"WARNING: Required column '{r}' not found. Candidates: {candidates}")

# 5) Clean and parse the 'release_date' column
# This replaces the 'clean_money_col' logic
if 'release_date' in df.columns:
    # Convert to datetime, handling the specific format "Month Day, Year"
    df['release_date_parsed'] = pd.to_datetime(df['release_date'].str.strip(), format='%B %d, %Y', errors='coerce')

    n_null = df['release_date_parsed'].isna().sum()
    print(f"Column 'release_date': converted to datetime as 'release_date_parsed' ({n_null} NaN after conversion)")

    # Extract year, which is useful for analysis
    df['release_year'] = df['release_date_parsed'].dt.year

    # Drop rows where the date is essential and couldn't be parsed
    before = len(df)
    df.dropna(subset=['release_date_parsed'], inplace=True)
    print(f"Dropped {before - len(df)} rows lacking valid release_date_parsed values")
else:
    print("WARNING: 'release_date' column not found. Skipping date parsing.")

# 6) Clean and parse the 'duration' column
# This is a new custom function, replacing 'clean_money_col'
if 'duration' in df.columns:
    print("Parsing 'duration' column...")
    # Extract numeric part
    df['duration_value'] = df['duration'].str.extract(r'(\d+)').astype(float)
    # Extract text part
    df['duration_unit'] = df['duration'].str.extract(r' (.*)')

    # Standardize units
    df['duration_unit'] = df['duration_unit'].replace({'Seasons': 'Season'})
    print("Created 'duration_value' and 'duration_unit' columns")
else:
    print("No 'duration' column found.")

# 7) Standardize 'country' column (replaces 'neighbourhood_group')
if 'country' in df.columns:
    # Fill missing values
    df['country'] = df['country'].fillna('Unknown').astype(str)
    # Extract just the *first* country listed
    df['primary_country'] = df['country'].str.split(',').str[0].str.strip()
    print("Standardized 'country' column and extracted 'primary_country'")
else:
    print("No 'country' column found.")

# 8) Fill other common missing values
for col in ['director', 'cast']:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')
        print(f"Filled missing values in '{col}' with 'Unknown'")

# 9) Show head and dtypes summary for inspection
print("\n--- Sample rows after cleaning ---")
print(df.head(5))
print("\n--- dtypes after cleaning ---")
print(df.dtypes)

# 10) Show value counts for key categorical columns (replaces 'room_type')
if 'media_type' in df.columns:
    pts = df['media_type'].value_counts().reset_index()
    print(f"\nContent types (media_type):\n{pts}")

if 'rating' in df.columns:
    pts = df['rating'].value_counts().reset_index().head(10)
    print(f"\nTop 10 Ratings (rating):\n{pts}")

if 'primary_country' in df.columns:
    pts = df['primary_country'].value_counts().reset_index().head(10)
    print(f"\nTop 10 Countries (primary_country):\n{pts}")

print("\n--- Cleaning pipeline complete. ---")
```

```python
import pandas as pd
import numpy as np

# --- 1. Load Real Netflix Data ---
# We must first load your actual file.
try:
    df = pd.read_csv("Netflix Dataset.csv")
    print("Loaded 'Netflix Dataset.csv'\n")
except FileNotFoundError:
    print("Error: 'Netflix Dataset.csv' not found.")
    df = pd.DataFrame(columns=['Category', 'Duration', 'Rating']) # Dummy
except NameError:
    # This handles cases in notebooks where 'df' might already be loaded
    print("Assuming 'df' is already loaded in your environment.\n")
    pass

# --- 2. Remove Sample Data ---
# This part of your script is not needed.
# data = {'neighbourhood group': [...], 'room type': [...], ...}
# df = pd.DataFrame(data)


# --- 3. Data Preparation (The Required Change) ---
# We must clean the mixed 'Duration' column before we can average it.

# 3a. Process Movies
movies_df = df[df['Category'] == 'Movie'].copy()
movies_df['duration_numeric'] = pd.to_numeric(
    movies_df['Duration'].str.replace(' min', ''),
    errors='coerce'
)

# 3b. Process TV Shows
shows_df = df[df['Category'] == 'TV Show'].copy()
# Handle both "1 Season" and "4 Seasons"
shows_df['duration_numeric'] = pd.to_numeric(
    shows_df['Duration'].str.replace(' Seasons', '').str.replace(' Season', ''),
    errors='coerce'
)

# 3c. Combine them back into one DataFrame
clean_df = pd.concat([movies_df, shows_df])

# 3d. Drop any rows where parsing failed
clean_df.dropna(subset=['duration_numeric', 'Rating', 'Category'], inplace=True)
print("Cleaned 'Duration' into 'duration_numeric' for Movies (min) and TV Shows (seasons).\n")


# --- 4. Run Your Adapted Analysis Code ---
# This replaces:
# arrn = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean().to_frame()

print("--- Average Duration by Category and Rating ---")

# We group by Category (replaces 'neighbourhood group')
# and Rating (replaces 'room type')
# and average our new 'duration_numeric' column (replaces 'review rate number')
avg_duration_by_group = clean_df.groupby(
    ['Category', 'Rating']
)['duration_numeric'].mean().to_frame(name='average_duration (min/seasons)')

# Sort to show highest averages first
avg_duration_by_group = avg_duration_by_group.sort_values(by='average_duration (min/seasons)', ascending=False)

print(avg_duration_by_group.head(15))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Load Real Netflix Data ---
try:
    df = pd.read_csv("Netflix Dataset.csv")
    print("Loaded 'Netflix Dataset.csv'")
except FileNotFoundError:
    print("Error: 'Netflix Dataset.csv' not found.")
    df = pd.DataFrame(columns=['Rating']) # Dummy for safety
except NameError:
    print("Assuming 'df' is already loaded in your environment.")
    pass

print("Generating plot for content distribution by rating...")

# --- 2. Create Value Counts DataFrame (Adapted) ---
# Replaced 'neighbourhood group' with 'Rating'
# Added .fillna() to handle missing ratings
# Added .nlargest(15) to get the Top 15 ratings for a clean chart
rating_counts = (
    df['Rating']
    .fillna('Not Rated')
    .value_counts()
    .nlargest(15)
    .to_frame(name='count')
)

# --- 3. Plot (Adapted) ---
plt.figure(figsize=(10, 6)) # Added a figure size
# Changed variable names and color
rating_bar = plt.bar(
    rating_counts.index,
    rating_counts['count'],
    color='firebrick',
    edgecolor='black'
)

# --- 4. Add Labels (Adapted) ---
# This code is adapted to use the new variable names
try:
    plt.bar_label(rating_bar, labels=rating_counts['count'], padding=4)
except AttributeError:
    print("Note: Using fallback for bar labels. Update Matplotlib (3.4+) for a cleaner look.")
    for bar in rating_bar:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 10,
                 f"{int(yval)}", ha='center', va='bottom')

# --- 5. Chart Formatting (Adapted) ---
# CRITICAL: Removed plt.ylim([0, 40000])
# This line would make your Netflix plot invisible.

# Changed labels
plt.xlabel('Rating (Top 15)', fontsize=11)
plt.ylabel('Number of Titles', fontsize=11)
plt.xticks(rotation=45, ha='right') # Keep rotation

# Changed title
plt.title('Content Distribution by Rating (Top 15)', fontsize=13)

plt.tight_layout() # Good practice
plt.show()

print("Plot displayed.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings

# Suppress warnings
warnings.filterwarnings('ignore')

# --- 1. Define constants (Adapted for Netflix) ---
FILE_NAME = 'Netflix Dataset.csv'
TARGET_COL = 'Category' # Changed from 'room type'

try:
    # --- 2. Load the dataset (Adapted for CSV) ---
    # Changed from pd.read_excel() to pd.read_csv()
    df = pd.read_csv(FILE_NAME)

    # --- 3. Clean column names (Your code, kept as-is) ---
    # This will turn 'Category' into 'Category' (no change)
    # but is good practice.
    df.columns = df.columns.str.strip()

    # --- 4. Check for adapted column ---
    if TARGET_COL not in df.columns:
        print(f"Error: Required column ('{TARGET_COL}') not found in the file.")
    else:
        # --- 5. Count listings (Adapted logic) ---
        # This will now count 'Movie' and 'TV Show'
        content_counts = df[TARGET_COL].value_counts()

        labels = content_counts.index
        sizes = content_counts.values

        # Netflix-themed colors
        colors = ['firebrick', 'gray']

        # Explode the largest slice (your code, kept as-is)
        explode = [0.1 if i == sizes.argmax() else 0 for i in range(len(sizes))]

        # --- 6. Plotting (Your code, adapted labels) ---
        fig, ax = plt.subplots(figsize=(10, 8))
        wedges, texts, autotexts = ax.pie(
            sizes,
            labels=labels,
            autopct='%1.1f%%',
            startangle=90,
            shadow=True,
            explode=explode,
            pctdistance=0.8,
            colors=colors  # Added colors
        )

        ax.axis('equal')  # Make pie chart circular

        # --- 7. Title (Adapted) ---
        ax.set_title(
            'Distribution of Netflix Content (Movie vs. TV Show)',
            fontsize=16,
            fontweight='bold',
            pad=20
        )

        # Style the percentage texts (Your code, kept as-is)
        for autotext in autotexts:
            autotext.set_color('white')
            autotext.set_weight('bold')

        # --- 8. Legend (Adapted) ---
        ax.legend(wedges, labels,
                  title="Content Type", # Changed title
                  loc="center left",
                  bbox_to_anchor=(1, 0, 0.5, 1))

        plt.tight_layout()
        plt.show()

# --- 9. Error Handling (Adapted error message) ---
except FileNotFoundError:
    print(f"Error: The file '{FILE_NAME}' was not found. Please check the file name.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
